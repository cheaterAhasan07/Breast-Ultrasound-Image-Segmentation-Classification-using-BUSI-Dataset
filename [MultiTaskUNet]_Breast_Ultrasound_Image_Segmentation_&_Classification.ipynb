```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
from torch.utils.data import DataLoader, Dataset
import os
import numpy as np
import matplotlib.pyplot as plt
from google.colab import drive
from tqdm import tqdm
from PIL import Image
from skimage.io import imread
from skimage import img_as_float
from scipy import signal
from numpy import pad
from sklearn.metrics import precision_recall_fscore_support

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

drive.mount('/content/drive')


print("\nContents of /content/drive/MyDrive:")
try:
    print(os.listdir("/content/drive/MyDrive"))
except Exception as e:
    print(f"Error listing drive contents: {e}")


correct_path_found = False
while not correct_path_found:
    dataset_path = input("\nPlease enter the full path of dataset folder (or 'quit' to exit): ")
    if dataset_path.lower() == 'quit':
        raise FileNotFoundError("Update path in the code.")
    if os.path.exists(dataset_path):
        print(f"\nFound dataset at: {dataset_path}")
        print(f"Contents of dataset folder: {os.listdir(dataset_path)}")
        correct_path_found = True
    else:
        print(f"\nPath {dataset_path} not found.")

image_dirs = [
    os.path.join(dataset_path, 'benign'),
    os.path.join(dataset_path, 'malignant'),
    os.path.join(dataset_path, 'normal')
]


for directory in image_dirs:
    if os.path.exists(directory):
        print(f"Found directory: {directory}")
        img_count = len([f for f in os.listdir(directory) if f.endswith('.png') and '_mask' not in f])
        mask_count = len([f for f in os.listdir(directory) if f.endswith('_mask.png')])
        print(f"  Contains {img_count} images and {mask_count} masks")
        if img_count > 0:
            sample_img = os.path.join(directory, [f for f in os.listdir(directory) if f.endswith('.png') and '_mask' not in f][0])
            img = img_as_float(imread(sample_img))
            if img.ndim == 2 or (img.ndim == 3 and img.shape[2] == 1):
                print(f"Sample image {sample_img} is grayscale with shape {img.shape}")
            else:
                print(f"Sample image {sample_img} is not grayscale, converting to grayscale")
                img = img_as_float(np.mean(img, axis=2))
            if mask_count > 0:
                sample_mask = os.path.join(directory, [f for f in os.listdir(directory) if f.endswith('_mask.png')][0])
                mask = img_as_float(imread(sample_mask))
                if mask.ndim == 2:
                    unique_vals = np.unique(mask)
                    print(f"Sample mask {sample_mask} has values {unique_vals}")
                else:
                    print(f"Sample mask {sample_mask} is not 2D")
    else:
        print(f"Directory {directory} not found")


def gaussian_kernel(width=1, size=5):
    x = np.linspace(-size // 2, size // 2, size)
    gaussian1d = np.exp(-0.5 * (x / width) ** 2)
    gaussian1d /= np.sum(gaussian1d)
    kernel = np.outer(gaussian1d, gaussian1d)
    return kernel / np.sum(kernel)

def apply_filter(image, kernel):
    pad_width = ((kernel.shape[0] // 2, kernel.shape[0] // 2),
                 (kernel.shape[1] // 2, kernel.shape[1] // 2))
    padded_img = pad(image, pad_width, mode='reflect')
    filtered = signal.convolve2d(padded_img, kernel, mode='valid', boundary='symm')
    return filtered

def resize_image(image, target_size=(256, 256)):
    h, w = image.shape
    new_h, new_w = target_size

    i_coords = np.linspace(0, h - 1, new_h, dtype=np.float32)
    j_coords = np.linspace(0, w - 1, new_w, dtype=np.float32)


    i_indices = np.round(i_coords).astype(np.int32)
    j_indices = np.round(j_coords).astype(np.int32)

    i_indices = np.clip(i_indices, 0, h - 1)
    j_indices = np.clip(j_indices, 0, w - 1)

    resized = image[i_indices][:, j_indices]

    return resized


def normalize_image(image):
    return image / image.max() if image.max() != 0 else image


class BreastUltrasoundDataset(Dataset):
    def __init__(self, image_dirs, transform=None):
        self.transform = transform
        self.samples = []
        for label_idx, label_dir in enumerate(image_dirs):
            if not os.path.exists(label_dir):
                print(f"Directory {label_dir} not found")
                continue
            try:
                images = [f for f in sorted(os.listdir(label_dir))
                         if f.endswith('.png') and '_mask' not in f]
                for img_name in images:
                    img_path = os.path.join(label_dir, img_name)
                    mask_path = os.path.join(label_dir, img_name.replace('.png', '_mask.png'))
                    if os.path.exists(mask_path):
                        mask = img_as_float(imread(mask_path))
                        if mask.ndim == 3:
                            mask = img_as_float(np.mean(mask, axis=2))
                        if np.max(mask) > 0:
                            self.samples.append((img_path, mask_path, label_idx))
                        else:
                            print(f"Skipping sample with empty mask: {mask_path}")
                    else:
                        print(f"Missing mask for {img_path}")
            except Exception as e:
                print(f"Error processing {label_dir}: {str(e)}")
        if not self.samples:
            raise RuntimeError("No valid image-mask pairs found in dataset directories")
        print(f"\nDataset successfully initialized with {len(self.samples)} valid samples (BUSI dataset with speckle noise)")

    def __len__(self):
        return len(self.samples)

    def __getitem__(self, idx):
        img_path, mask_path, label = self.samples[idx]
        try:
            img = img_as_float(imread(img_path))
            if img.ndim == 3:
                img = img_as_float(np.mean(img, axis=2))
            mask = img_as_float(imread(mask_path))
            if mask.ndim == 3:
                mask = img_as_float(np.mean(mask, axis=2))
            gaussian_kern = gaussian_kernel(width=1, size=5)
            use_gaussian = True
            if use_gaussian:
                img = apply_filter(img, gaussian_kern)
            img = resize_image(img, target_size=(256, 256))
            img = normalize_image(img)
            # Binarize mask
            mask = (mask > 0.5).astype(np.float32)
            mask = resize_image(mask, target_size=(256, 256))
            if not np.any(mask):
                raise ValueError(f"Empty mask after processing for {mask_path}")
            img_rgb = np.stack([img, img, img], axis=-1)
            mask = np.expand_dims(mask, axis=-1)
            sample = {
                'image': torch.tensor(img_rgb, dtype=torch.float32).permute(2, 0, 1),
                'mask': torch.tensor(mask, dtype=torch.float32).permute(2, 0, 1),
                'label': torch.tensor(label, dtype=torch.long)
            }
            if self.transform:
                sample = self.transform(sample)
            return sample
        except Exception as e:
            print(f"Error loading {img_path}: {str(e)}")
            return {
                'image': torch.zeros((3, 256, 256), dtype=torch.float32),
                'mask': torch.zeros((1, 256, 256), dtype=torch.float32),
                'label': torch.tensor(label, dtype=torch.long)
            }


class MultiTaskUNet(nn.Module):
    def __init__(self, in_channels=3, out_channels_seg=1, num_classes_class=3):
        super(MultiTaskUNet, self).__init__()
        self.enc1 = nn.Sequential(
            nn.Conv2d(in_channels, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU()
        )
        self.pool1 = nn.MaxPool2d(2, 2)
        self.enc2 = nn.Sequential(
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU()
        )
        self.pool2 = nn.MaxPool2d(2, 2)
        self.bottleneck = nn.Sequential(
            nn.Conv2d(128, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Conv2d(256, 256, 3, padding=1),
            nn.BatchNorm2d(256),
            nn.ReLU(),
            nn.Dropout(0.3)
        )
        self.upconv2 = nn.ConvTranspose2d(256, 128, 2, stride=2)
        self.dec2 = nn.Sequential(
            nn.Conv2d(256, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.Conv2d(128, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU()
        )
        self.upconv1 = nn.ConvTranspose2d(128, 64, 2, stride=2)
        self.dec1 = nn.Sequential(
            nn.Conv2d(128, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.Conv2d(64, out_channels_seg, 3, padding=1)
        )
        self.classifier = nn.Sequential(
            nn.AdaptiveAvgPool2d(1),
            nn.Flatten(),
            nn.Linear(256, 256),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(256, num_classes_class)
        )

    def forward(self, x):
        enc1 = self.enc1(x)
        pool1 = self.pool1(enc1)
        enc2 = self.enc2(pool1)
        pool2 = self.pool2(enc2)
        bottleneck = self.bottleneck(pool2)
        up2 = self.upconv2(bottleneck)
        dec2_input = torch.cat([up2, enc2], dim=1)
        dec2 = self.dec2(dec2_input)
        up1 = self.upconv1(dec2)
        dec1_input = torch.cat([up1, enc1], dim=1)
        seg_out = self.dec1(dec1_input)
        cls_out = self.classifier(bottleneck)
        return seg_out, cls_out


def dice_loss(pred, target, smooth=1e-5):
    pred = torch.sigmoid(pred)
    intersection = (pred * target).sum()
    return 1 - ((2. * intersection + smooth) / (pred.sum() + target.sum() + smooth))


def combined_loss(seg_pred, seg_true, cls_pred, cls_true, seg_weight=0.7, cls_weight=0.3):
    seg_loss = dice_loss(seg_pred, seg_true)
    cls_loss = F.cross_entropy(cls_pred, cls_true)
    total_loss = seg_weight * seg_loss + cls_weight * cls_loss
    return total_loss, seg_loss.item(), cls_loss.item()


def compute_iou(seg_pred, seg_true, threshold=0.5, smooth=1e-5):
    seg_pred_binary = (torch.sigmoid(seg_pred) > threshold).float()
    intersection = (seg_pred_binary * seg_true).sum()
    union = seg_pred_binary.sum() + seg_true.sum() - intersection
    iou = (intersection + smooth) / (union + smooth)
    return iou.item()

try:
    train_dataset = BreastUltrasoundDataset(image_dirs)
    if len(train_dataset) == 0:
        raise RuntimeError("Dataset is empty.")
    train_size = int(0.8 * len(train_dataset))
    val_size = len(train_dataset) - train_size
    train_dataset, val_dataset = torch.utils.data.random_split(train_dataset, [train_size, val_size])

    train_loader = DataLoader(
        train_dataset,
        batch_size=8,
        shuffle=True,
        num_workers=0,
        pin_memory=True if torch.cuda.is_available() else False
    )
    val_loader = DataLoader(
        val_dataset,
        batch_size=8,
        shuffle=False,
        num_workers=0,
        pin_memory=True if torch.cuda.is_available() else False
    )
    print(f"\nDataLoaders successfully created - {len(train_dataset)} training samples, {len(val_dataset)} validation samples")
except Exception as e:
    print(f"\nError creating dataset: {str(e)}")
    raise

try:
    sample = train_dataset[0]
    print("\nSample verification:")
    print(f"Image shape: {sample['image'].shape}")
    print(f"Mask shape: {sample['mask'].shape}")
    print(f"Label: {sample['label']}")
    plt.figure(figsize=(12, 4))
    plt.subplot(1, 2, 1)
    plt.imshow(sample['image'].permute(1, 2, 0).numpy())
    plt.title("Sample Image")
    plt.subplot(1, 2, 2)
    plt.imshow(sample['mask'].permute(1, 2, 0).numpy()[:,:,0], cmap='gray')
    plt.title("Sample Mask")
    plt.savefig('sample_visualization.png')
except Exception as e:
    print(f"Error verifying sample: {str(e)}")

model = MultiTaskUNet().to(device)
print("\nModel initialized:")
print(model)


optimizer = optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-5)
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=3, verbose=True)


def evaluate(model, loader, threshold=0.5, return_metrics=False):
    model.eval()
    total_loss = 0.0
    total_seg_loss = 0.0
    total_cls_loss = 0.0
    total_dice = 0.0
    total_iou = 0.0
    total_pixel_acc = 0.0
    total_correct = 0
    total_samples = 0
    all_preds = []
    all_labels = []
    smooth = 1e-5

    with torch.no_grad():
        for batch in tqdm(loader, desc="Evaluating"):
            if batch is None:
                continue
            images = batch['image'].to(device)
            masks = batch['mask'].to(device)
            labels = batch['label'].to(device)
            seg_pred, cls_pred = model(images)
            loss, seg_loss, cls_loss = combined_loss(seg_pred, masks, cls_pred, labels)
            total_loss += loss.item()
            total_seg_loss += seg_loss
            total_cls_loss += cls_loss

            seg_pred_binary = (torch.sigmoid(seg_pred) > threshold).float()
            intersection = (seg_pred_binary * masks).sum()
            union = seg_pred_binary.sum() + masks.sum() - intersection
            iou = (intersection + smooth) / (union + smooth)
            total_iou += iou.item()

            dice = (2. * intersection + smooth) / (seg_pred_binary.sum() + masks.sum() + smooth)
            total_dice += dice.item()

            pixel_acc = (seg_pred_binary == masks).float().mean()
            total_pixel_acc += pixel_acc.item()

            cls_pred_labels = cls_pred.argmax(dim=1)
            total_correct += (cls_pred_labels == labels).sum().item()
            total_samples += labels.size(0)
            all_preds.extend(cls_pred_labels.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    avg_loss = total_loss / len(loader)
    avg_seg = total_seg_loss / len(loader)
    avg_cls = total_cls_loss / len(loader)
    avg_dice = total_dice / len(loader)
    avg_iou = total_iou / len(loader)
    avg_pixel_acc = total_pixel_acc / len(loader)
    avg_acc = total_correct / total_samples
    precision, recall, f1, _ = precision_recall_fscore_support(all_labels, all_preds, average='weighted', zero_division=0)

    if return_metrics:
        return avg_loss, avg_seg, avg_cls, avg_dice, avg_iou, avg_pixel_acc, avg_acc, precision, recall, f1
    return avg_loss

#task3
def train_model(model, train_loader, val_loader, optimizer, scheduler, epochs=20, threshold=0.5):
    model.train()
    history = {
        'train_loss': [], 'train_seg_loss': [], 'train_cls_loss': [],
        'val_loss': [], 'val_seg_loss': [], 'val_cls_loss': [],
        'val_dice': [], 'val_iou': [], 'val_pixel_acc': [], 'val_acc': [],
        'val_precision': [], 'val_recall': [], 'val_f1': []
    }
    best_val_iou = 0.0
    for epoch in range(epochs):
        model.train()
        epoch_loss = 0.0
        epoch_seg_loss = 0.0
        epoch_cls_loss = 0.0
        valid_batches = 0
        pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")
        for batch in pbar:
            if batch is None:
                continue
            images = batch['image'].to(device)
            masks = batch['mask'].to(device)
            labels = batch['label'].to(device)
            seg_pred, cls_pred = model(images)
            loss, seg_loss, cls_loss = combined_loss(seg_pred, masks, cls_pred, labels)
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
            valid_batches += 1
            epoch_loss += loss.item()
            epoch_seg_loss += seg_loss
            epoch_cls_loss += cls_loss
            pbar.set_postfix({
                'loss': loss.item(),
                'seg': seg_loss,
                'cls': cls_loss
            })


            if epoch == 0 and valid_batches == 1:
                with torch.no_grad():
                    seg_pred_binary = (torch.sigmoid(seg_pred) > threshold).float()
                    iou = compute_iou(seg_pred, masks, threshold=threshold)
                    print(f"Manual IoU for first batch (epoch {epoch+1}): {iou:.4f}")
                    plt.figure(figsize=(15, 5))
                    plt.subplot(1, 3, 1)
                    plt.imshow(images[0].cpu().permute(1, 2, 0).numpy())
                    plt.title("Input Image")
                    plt.subplot(1, 3, 2)
                    plt.imshow(masks[0].cpu().permute(1, 2, 0).numpy()[:,:,0], cmap='gray')
                    plt.title("Ground Truth Mask")
                    plt.subplot(1, 3, 3)
                    plt.imshow(seg_pred_binary[0].cpu().permute(1, 2, 0).numpy()[:,:,0], cmap='gray')
                    plt.title(f"Predicted Mask (IoU: {iou:.4f})")
                    plt.savefig('epoch_0_prediction.png')
                    plt.close()

        if valid_batches == 0:
            print(f"Epoch {epoch+1} had no valid batches!")
            continue
        avg_loss = epoch_loss / valid_batches
        avg_seg = epoch_seg_loss / valid_batches
        avg_cls = epoch_cls_loss / valid_batches
        history['train_loss'].append(avg_loss)
        history['train_seg_loss'].append(avg_seg)
        history['train_cls_loss'].append(avg_cls)
        val_metrics = evaluate(model, val_loader, threshold=threshold, return_metrics=True)
        val_loss, val_seg, val_cls, val_dice, val_iou, val_pixel_acc, val_acc, val_precision, val_recall, val_f1 = val_metrics
        history['val_loss'].append(val_loss)
        history['val_seg_loss'].append(val_seg)
        history['val_cls_loss'].append(val_cls)
        history['val_dice'].append(val_dice)
        history['val_iou'].append(val_iou)
        history['val_pixel_acc'].append(val_pixel_acc)
        history['val_acc'].append(val_acc)
        history['val_precision'].append(val_precision)
        history['val_recall'].append(val_recall)
        history['val_f1'].append(val_f1)
        scheduler.step(val_loss)
        print(f"\nEpoch {epoch+1} Summary:")
        print(f"Train Loss: {avg_loss:.4f} | Train Seg: {avg_seg:.4f} | Train Cls: {avg_cls:.4f}")
        print(f"Val Loss: {val_loss:.4f} | Val Seg: {val_seg:.4f} | Val Cls: {val_cls:.4f}")
        print(f"Val Dice: {val_dice:.4f} | Val IoU: {val_iou:.4f} | Val Pixel Acc: {val_pixel_acc:.4f}")
        print(f"Val Acc: {val_acc:.4f} | Val Precision: {val_precision:.4f} | Val Recall: {val_recall:.4f} | Val F1: {val_f1:.4f}")
        if val_iou > best_val_iou:
            best_val_iou = val_iou
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'best_iou': best_val_iou,
            }, 'best_model.pth')
            print(f"model saved with IoU score: {best_val_iou:.4f}")
    return history


train_model(model, train_loader, val_loader, optimizer, scheduler, epochs=20, threshold=0.5)
torch.save({
    'model_state_dict': model.state_dict(),
    'optimizer_state_dict': optimizer.state_dict(),
}, 'iiye lo breast_ultrasound_final_model.pth')
print("\nFinal model saved to 'iiye lo breast_ultrasound_final_model.pth'\n")
print("\nTraining and evaluation complete!")
```
